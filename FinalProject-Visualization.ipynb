```python
import pandas as pd

# Load the CSV file again with the correct delimiter
df_electricity = pd.read_csv("./My_Electricity_Consumption.csv", delimiter=';')
# df_price = pd.read_csv("./My_sahkon-hinta-010121-231024.csv")

# Rename the columns for easier access and clean-up
df_electricity.columns = [
    'Time', 'Energy (kWh)', 'Status', 'Temperature']

display(df_electricity.head())

df_electricity['Energy (kWh)'] = df_electricity['Energy (kWh)'].str.replace(
    ',', '.').astype(float)
df_electricity['Temperature'] = df_electricity['Temperature'].str.replace(
    ',', '.').astype(float)

display(df_electricity.head())

# Convert 'Time' column to datetime
df_electricity['Time'] = pd.to_datetime(pd.to_datetime(
    df_electricity['Time'].str.strip(), format='%d.%m.%Y %H:%M', errors='coerce'))

display(df_electricity.head())
```

```python
# price file read and operations
# Load the CSV file again with the correct delimiter
df_price = pd.read_csv("./My_sahkon-hinta-010121-231024.csv", delimiter=';')

# Rename the columns for easier access and clean-up
df_price.columns = ['Time', 'Price (cent/kWh)']

display(df_price.head())

# Convert 'Time' column to datetime
df_price['Time'] = pd.to_datetime(df_price['Time'], format='%d/%m/%Y %H.%M')
display(df_price.head())

df_price['Price (cent/kWh)'] = df_price['Price (cent/kWh)'].str.replace(
    ',', '.').astype(float)

display(df_price.head())

# Merge the data frames on the 'Time' column
df_merged = pd.merge(df_electricity, df_price, on='Time', how='outer',)

print('df_merged')
print(df_merged)
```

```python
# # Calculate the hourly bill
df_merged['Hourly Bill (€)'] = (df_merged['Energy (kWh)'] *
                                (df_merged['Price (cent/kWh)'] / 100)).round(2)


# Convert 'Time' column to datetime for proper grouping
df_merged['Time'] = pd.to_datetime(df_merged['Time'])

print(df_merged)
```

```python
from datetime import datetime


start_time = datetime.strptime("2022/01/01", "%Y/%m/%d")
end_time = datetime.strptime("2024/06/01", "%Y/%m/%d")

# total consumption over period

df_filtered = df_merged[(df_merged['Time'] >= start_time)
                        & (df_merged['Time'] <= end_time)]

total_consumption = df_filtered['Energy (kWh)'].sum()

total_bill = df_filtered['Hourly Bill (€)'].sum()

# Display the output
print(f"Start time\n{start_time.strftime('%Y/%m/%d')}")
print(f"\nEnd time\n{end_time.strftime('%Y/%m/%d')}")
print(f"\nShowing range: {start_time.strftime(
    '%Y-%m-%d')} - {end_time.strftime('%Y-%m-%d')}")
print(f"Total consumption over the period: {total_consumption} kWh")
print(f"Total bill over the period: {total_bill} €")
```
